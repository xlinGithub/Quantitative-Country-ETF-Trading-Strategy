```python
#import the working library
import pandas as pd
import datetime
import os
```

```python
#read mapping values as specified in excel 
mapping = pd.read_excel('mapping.xlsx')
```

```python
#show mapping result
mapping.head()
```

```python
#specified raw data file location
raw_data_file_name = 'raw data\BMB_Countries Updated 17.3.xlsx'
```

# Fundemental Factors

```python
#read row data file's equity fundamental sheet
raw_factor_data = pd.read_excel(raw_data_file_name, sheet_name='Equity Fundamental') # takes some time
```

```python
#specified number column and read the indexes for each tickers and save result by sorted date and factor data begins in 1990, Jan, 1
n_col = 5
for i in range(int(raw_factor_data.shape[1]/n_col)):
    index = raw_factor_data.columns[i*n_col+1][:-6]
    ticker = mapping.set_index('Underlying or Proxy Index').loc[index, 'Ticker']
    combined_factor_data = pd.DataFrame(raw_factor_data[raw_factor_data.columns[i*n_col:(i+1)*n_col]].iloc[5:].values, columns=['Date', 'pe', 'pcf', 'div_yield', 'pb'])
    try:
        combined_factor_data['Date'] = [datetime.datetime.fromordinal(datetime.datetime(1900, 1, 1).toordinal() + d - 2).strftime('%Y-%m-%d') for d in combined_factor_data['Date']]
    except:
        pass
    combined_factor_data = combined_factor_data.set_index('Date').dropna().sort_values('Date')
    if combined_factor_data.shape[0] > 1: 
        for col in combined_factor_data.columns:
            combined_factor_data[[col]].to_csv('factor data/'+col+'/'+ticker+'.csv')
```

# Factor Data Availability Check

```python
#save ticker, country, market by ticker 
data_availability_df = mapping[['Ticker', 'Country', 'Market']].set_index('Ticker')
factor_list = [f[1] for f in os.walk('factor data')][0]
```

```python
#check all factor and save whether we obtain all specified in each etf
for i in data_availability_df.index:
    for f in factor_list:
        data_availability_df.loc[i, f] = os.path.exists('factor data/'+f+'/'+i+'.csv')
```

```python
#function to sort each factor as according to dataframe's availaiblity in each factor order
def highlight_max(s):
    '''
    highlight the maximum in a Series yellow.
    '''
    is_max = s == s.max()
    return ['background-color: lightgreen' if v else '' for v in is_max]

#show the sorted data availability dataframe
data_availability_df[data_availability_df.columns[2:]].style.apply(highlight_max)
```
